```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("theta_definitions_comparisons_3.csv")
df["ACR"] = df["C"] / df["T"]
#df = df[(df["theta_operator"] == "ALL")  ]
df = df[(df["theta_definition"] == "VAR")  ]

df["theta_b"] = df["theta_definition"] + " " + df["theta_operator"]
df
```

```python
df2 = df.groupby(["budget","theta_b"]).mean().reset_index()
df2["budget"] = df2["budget"] // 4
df2["budget"] = df2["budget"] * 4
df2 = df2.groupby(["budget","theta_b"]).mean().reset_index()
```

```python
ax = sns.scatterplot(data=df2, x="T", y="ACR", hue = "theta_b", size = "budget")
```
