```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("theta_definitions_comparisons_with_cardinals_4.csv")
df["ACR"] = df["C"] / df["T"]
#df = df[(df["theta_operator"] == "ALL") | (df["theta_operator"] == "SVM") | (df["theta_operator"] == "LPM")   ]
#df = df[(df["theta_definition"] == "LEX2")  ]
df["theta_definition"] = df["theta_definition"].apply(lambda x: "EXP" if x == "VAR" else x)
df["theta_definition"] = df["theta_definition"].apply(lambda x: "CARD" if x == "LPM" else x)

df["theta_b"] =+ df["theta_operator"]
np.unique(df.budget.values)
```

```python
df2 = df.groupby(["budget","theta_b"]).mean().reset_index()
#df2 = df2[df2.budget == df2.budget.values.max()]
df2["budget"] = df2["budget"] // 5
df2["budget"] = df2["budget"] * 5
df2 = df2.groupby(["budget","theta_b"]).mean().reset_index()
df2.budget
```

```python
ax = sns.scatterplot(data=df2[df2.budget == 30], x="T", y="ACR", hue = "theta_b")
plt.savefig("theta_definitions_ACR.png")
```

```python
df = pd.read_csv("theta_definitions_comparisons_3.csv")
df["ACR"] = df["C"] / df["T"]
#df = df[(df["theta_operator"] == "ALL")  ]
df = df[(df["theta_definition"] == "LEX2")  ]
df["theta_definition"] = df["theta_definition"].apply(lambda x: "EXP" if x == "VAR" else x)

df["theta_b"] = df["theta_definition"] + " " + df["theta_operator"]
df2 = df.groupby(["budget","theta_b"]).mean().reset_index()
#df2 = df2[df2.budget == df2.budget.values.max()]

df2["budget"] = df2["budget"] // 3
df2["budget"] = df2["budget"] * 3
df2 = df2.groupby(["budget","theta_b"]).mean().reset_index()

ax = sns.scatterplot(data=df2, x="T", y="ACR", hue = "theta_b", size = "budget")
plt.savefig("theta_definitions_ACR_2.png")
```
